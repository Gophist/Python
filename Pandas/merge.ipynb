## pd.merge()

合并
https://blog.csdn.net/brucewong0516/article/details/82707492

````python
import pandas as pd
````

````python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
df1
````

````python
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})
df2
````

```py
  pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None,
  left_index=False, right_index=False, sort=True,
  suffixes=('_x', '_y'), copy=True, indicator=False,
  validate=None)
```

````python
re = pd.merge(df2, left_on='lkey', right_on='rkey', suffixes=('_l', '_r'))
re
````

````python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                       'A': ['A0', 'A1', 'A2', 'A3'],
                       'B': ['B0', 'B1', 'B2', 'B3']})
left
````

````python
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']})
right
````

````python
result = pd.merge(left, right, on='key')

# on参数传递的key作为连接键
result
````

传入 `on` 的参数是列表

````python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                      'key2': ['K0', 'K1', 'K0', 'K1'],
                         'A': ['A0', 'A1', 'A2', 'A3'],
                         'B': ['B0', 'B1', 'B2', 'B3']})
left
````

````python
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                         'C': ['C0', 'C1', 'C2', 'C3'],
                         'D': ['D0', 'D1', 'D2', 'D3']})
right
````

````python
result = pd.merge(left, right, on=['key1', 'key2'])
result
````

````python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                     'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']})
left
````

````python
right = pd.DataFrame({'key': ['K0', 'K1'], 'C': ['C0', 'C1']})
right
````

````python
# how: 指定两个子DataFrame的连接方式，
# 默认内连接删去了key列值不相同的行。
re = pd.merge(left, right, on=["key"])
re
````
